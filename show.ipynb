```python
import json
```

```python
def show(n=5):
    name_a = "gemini-2b-contrastive"
    name_b = "gemini-2b-contrastive-reward-1-5-1.5"
    data_a = json.load(open("biogen/{}.json".format(name_a)))
    data_b = json.load(open("biogen/{}.json".format(name_b)))
    
    data_a = list(data_a.values())[1:]
    data_b = list(data_b.values())[1:]

    start_idx = 0
    for i in range(start_idx, start_idx+n):
        print(data_a[i]["name"])
        print("-" * 100)
        print("\n\n".join(data_a[i]["contexts"][:3]))
        print("-" * 100)
        print(data_a[i]["response"])
        print("-" * 100)
        print(data_b[i]["response"])
        print("-" * 100)
        print("\n" * 5)
```

```python
show(n=10)
```
